### Domain and Dataset

Domain: We are interested in studying the salary levels of data science and data related professions. More specifically, we strive to answer the question: what factors influence the salary level (high vs. low) of job listings in the data science profession? Is it the title (e.g. Senior Data Scientist, Data Engineer, etc.)? How about the job description (such as, "involved in machine learning")?

Dataset: Our dataset consists of job listings from indeed.com. Each individual job listing has location (city), title, summary (something akin to job description), and a given salary range for that specific job position. Derived from this salary range is a parsed salary, which is the mean of the lower and upper numbers of the salary range.

### Problem Statement

Our goal is to come up with a model that can accurately predict whether a job listing will be a high salary or low salary job. We will then be able to use this model to inform our clients as to whether a certain job (that of course does not have salary information listed) will be a high paying or lower paying job.

### Data Preparation

We will begin by loading, cleaning, and  preparing our data for analysis:

```python
# Load modules
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataframe
indeed_df = pd.read_csv('./assets/indeed-scraped-job-postings.csv')
indeed_df.head(5)
```

```python
# Check for null values
indeed_df.isnull().sum()
```

```python
# Inspect values where parsed_salary is NaN. Because salaries are listed on a per day
# basis, I do not believe imputing these NaN values (by city average, for example) is a
# good idea. Also, there are only 7 entries with NaN values (out of a total of 413), so we
# can drop these entries.
indeed_df_parsed_salary_nan = indeed_df[indeed_df['parsed_salary'].isnull()]
indeed_df_parsed_salary_nan
```

```python
# Drop na values. Only our parsed_salary column has na values, so we will run .dropna on the
# whole dataframe
indeed_df.dropna(inplace=True)
```

```python
# Drop the salary column, which we will not be using
indeed_df = indeed_df.drop('salary', axis=1)
indeed_df.head(1)
```

```python
# Obtain measures of central tendency for parsed_salary
indeed_df.describe()
```

```python
# We will use the median of the parsed_salary as our dividing line. Anything above this
# median will be considered a high salary. Anything lower than this median will be considered
# a low salary. From our .describe() table above, we see the 50th percentile (50%) is
# 100,000. We will confirm this now:
indeed_df['parsed_salary'].median()
```

```python
# Make a new column denoting high salary (with a 1), or low salary (with a 0)
indeed_df['salary_level'] = indeed_df['parsed_salary'].apply(lambda x: 1 if x >= 100000.0 else 0)
```

```python
# Here, we get an exact number of how many "high" salaries vs "low" salaries there
# are in our dataset
indeed_df['salary_level'].value_counts()
```

```python
# import get_stop_words to remove stop words from our summary and titles
from stop_words import get_stop_words

stop_words = get_stop_words('en')

# remove unicode character in front of each word in stop_words list
stop_words_without_unicode_char = [str(x) for x in stop_words]

# functions to remove stop words and unnecessary punctuation
def remove_stop_words(column):
    cleaned_list = [x for x in column if x not in stop_words_without_unicode_char]
    return cleaned_list

def remove_punctuation(list_of_words):
    cleaned_punc_list = []
    for word in list_of_words:
        word = word.replace('.','')
        word = word.replace('..','')
        word = word.replace('...','')
        word = word.replace('....','')
        word = word.replace(',','')
        word = word.replace(':','')
        word = word.replace('/',' ')
        word = word.replace('*','')
        word = word.replace('(','')
        word = word.replace(')','')
        word = word.replace('!','')
        cleaned_punc_list.append(word)
    return cleaned_punc_list
```

```python
# Tokenize string in summary column
indeed_df['summary'] = indeed_df['summary'].str.lower()
indeed_df['summary_tokenized_items'] = indeed_df['summary'].str.split(' ')

# remove stop words and unnecessary punctuation from tokenized items, title column
indeed_df['summary_tokenized_stop_words_removed'] = indeed_df['summary_tokenized_items'].apply(remove_stop_words)
indeed_df['summary_tokenized_stop_words_removed'] = indeed_df['summary_tokenized_stop_words_removed'].apply(remove_punctuation)
```

```python
# Tokenize string in title column
indeed_df['title'] = indeed_df['title'].str.lower()
indeed_df['title_tokenized_items'] = indeed_df['title'].str.split(' ')

# remove stop words and unnecessary punctuation from tokenized items, title column
indeed_df['title_tokenized_stop_words_removed'] = indeed_df['title_tokenized_items'].apply(remove_stop_words)
indeed_df['title_tokenized_stop_words_removed'] = indeed_df['title_tokenized_stop_words_removed'].apply(remove_punctuation)
```

```python
# generate our features

# first, for words in the job summary
summary_words_dictionary = {}
for list_entry in indeed_df['summary_tokenized_stop_words_removed']:
    for word in list_entry:
        summary_words_dictionary[word] = summary_words_dictionary.get(word,0) + 1

for i in summary_words_dictionary.keys():
    indeed_df['sum_kw'+'_'+i] = indeed_df['summary'].apply(lambda x: 1 if i in x else 0)

# next, for words in the job title
title_words_dictionary = {}
for list_entry in indeed_df['title_tokenized_stop_words_removed']:
    for word in list_entry:
        title_words_dictionary[word] = title_words_dictionary.get(word,0) + 1
        
for i in title_words_dictionary.keys():
    indeed_df['title_kw'+'_'+i] = indeed_df['title'].apply(lambda x: 1 if i in x else 0)
```

### Solution Statement

Because we are interested in whether a certain job listing will be either a high or low salary, which is a categorical outcome/prediction, we will use logistic regression. In addition, we will also run a 10-fold cross validation (to examine how well our model will perform in the "real world") and a grid search.

### Metric

Our measure of success will be accuracy. How well does our model predict whether a job listing's salary will be high (greater than or equal to 100,000 per year) or low (less than 100,000 per year)? For example, if it guesses correctly 7 out of 10 times, the accuracy is 70%.

### Benchmark

Our benchmark will be the baseline accuracy, which is defined as:
The number of cases in our majority class / The total number of cases

From our EDA above, using value_counts(), we found that our majority class, high salary level job listings, occurred 209 times. The total number of job listings was 406. Therefore, our baseline accuracy is 51.5%:

```python
baseline_accuracy = 209*1./406
baseline_accuracy
```

```python
# Start logistic regression here
# import necessary tools from sklearn, statsmodels, metrics, etc.
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
# Drop the columns we will not be using for our logistic regression
indeed_df_final = indeed_df.drop(labels=['parsed_salary','city','title', 'summary','company',
                                        'summary_tokenized_items',
                                        'summary_tokenized_stop_words_removed',
                                        'title_tokenized_items',
                                        'title_tokenized_stop_words_removed'], axis=1)
```

```python
# Split our data into y and X
y = indeed_df_final['salary_level']
X = indeed_df_final.drop('salary_level', axis=1)
```

```python
# Logsitic Regression with Ridge, no cross validation
lr = LogisticRegression(penalty = 'l2', C=1.0)
lr.fit(X, y)
print ('INITIAL MODEL ACCURACY (Ridge regularization, NO Cross Validation: )'), lr.score(X, y)
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

from sklearn.metrics import accuracy_score
print('Accuracy: %.4f') % accuracy_score(y_test, y_pred)
```

```python
# Our confusion matrix (aka contingency table) for the train test split above:
from sklearn.metrics import confusion_matrix
confu_mtrx = confusion_matrix(y_true = y_test, y_pred = y_pred)

fig, ax = plt.subplots(figsize=(4, 4))
ax.matshow(confu_mtrx, cmap=plt.cm.Greens, alpha=0.3)
for i in range(confu_mtrx.shape[0]):
    for j in range(confu_mtrx.shape[1]):
        ax.text(x=j, y=i,
            s=confu_mtrx[i, j],
            va='center', ha='center')
plt.xlabel('predicted salary level')
plt.ylabel('true salary level')
print('Confusion Matrix for Type I and Type II Error')
plt.show()
```

Our confusion matrix shows the number of Type I and Type II errors.

In our table, a Type I error, or a False Positive, would be in the upper right quadrant: we incorrectly predicted 7 times that a job was high salary, when in fact, it was a low salary job. In statistical terms, we incorrectly rejected the null hypothesis (i.e. we should have accepted the null hypothesis).

As for Type II error, or False Negative, we see this in our lower left quadrant. We incorrectly predicted 17 times that a job was low salary, when in fact, it was a high salary job. To put this another way, we failed to reject a true null hypothesis.

```python
# Cross Validation with 10 folds
scores = cross_val_score(lr, X, y, scoring='accuracy', cv=10)

print ('CV 10-Fold Scores: ', scores)
print ''
print ('CV 10-Fold Mean Accuracy: ', scores.mean() )
```

```python
# GridSearch
from sklearn.model_selection import GridSearchCV
param_range = [0.0001, 0.001, 0.01, .1, 1.0, 10.0, 100.0, 1000.0]
param_grid = [{'C': param_range}]

gs = GridSearchCV(estimator=lr, param_grid=param_grid, scoring='accuracy', cv=10, n_jobs=-1)
gs = gs.fit(X_train, y_train)

print('Grid Search Best Score: %.4f' % gs.best_score_)
print('Grid Search Best Parameter for C: ')
print gs.best_params_
```

```python
from sklearn.metrics import roc_curve, auc

grid_search_lr = gs.best_estimator_

y_pp_lr = grid_search_lr.predict_proba(X)
fpr_lr, tpr_lr, _ = roc_curve(y, y_pp_lr[:, 1])
roc_auc_lr = auc(fpr_lr, tpr_lr)

plt.figure(figsize=[6,4])
plt.plot(fpr_lr, tpr_lr, label='logreg ROC curve (area = %0.2f)' % roc_auc_lr, 
         linewidth=4, color='darkred')
plt.plot([0, 1], [0, 1], 'k--', linewidth=4)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=16)
plt.ylabel('True Positive Rate', fontsize=16)
plt.title('Receiver operating characteristic curve', fontsize=20)
plt.show()
```
